```python
import matplotlib.pyplot as plt

# Dữ liệu
age = [18, 20, 22, 30, 34, 40, 60, 66, 70]
expenditure = [80, 90, 85, 50, 64, 60, 30, 40, 25]

# Vẽ biểu đồ
plt.figure(figsize=(8,5))
plt.plot(age, expenditure, marker='o', linestyle='-', color='blue', label='Expenditure')

# Gán nhãn
plt.title('Age vs Expenditure')
plt.xlabel('Age')
plt.ylabel('Expenditure')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Hiển thị đồ thị
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Dữ liệu (Age, Expenditure)
X = np.array([
    [18, 80],
    [20, 90],
    [22, 85],
    [30, 50],
    [34, 64],
    [40, 60],
    [60, 30],
    [66, 40],
    [70, 25]
])

# Thử phân cụm với K = 3
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Vẽ biểu đồ scatter
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, s=100, cmap='viridis', label='Data points')
plt.scatter(centers[:,0], centers[:,1], c='red', s=200, marker='X', label='Centroids')

plt.title('K-Means Clustering on Age vs Expenditure')
plt.xlabel('Age')
plt.ylabel('Expenditure')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
